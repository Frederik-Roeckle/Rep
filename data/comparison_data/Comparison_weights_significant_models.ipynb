```python
import pandas as pd

# Read the CSV file into a Pandas DataFrame
df1 = pd.read_csv('weights of significant forward model - new test dataset.csv', header=None).drop(0)

# Rename the columns
df1.columns = ['Description', 'weights of significant forward model - new test dataset']

print(df1.head())

# Replace '.' with ',' in the 'Number' column
#df['Value'] = df['Value'].str.replace('.', ',')

# Save the new DataFrame to an Excel file
#df.to_excel('output2.xlsx', index=False)


```

```python
# Read the CSV file into a Pandas DataFrame
df2 = pd.read_csv('weights of significant forward model - replicated dataset.csv', header=None).drop(0)

# Rename the columns
df2.columns = ['Description', 'weights of significant forward model - replicated dataset']

print(df2.head())

```

```python
data = pd.merge(df1, df2, on='Description', how = 'inner')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create the plot with a larger figure size
fig, ax = plt.subplots(figsize=(16, 10))  # Adjust the figsize parameter for a larger plot

# Calculate the positions for the bars
x = np.arange(len(data['Description']))  # the label locations
width = 0.35  # the width of the bars

# Plot the bars
bars1 = ax.bar(x - width/2, data['weights of significant forward model - new test dataset'], width, label='weights of significant forward model - new test dataset')
bars2 = ax.bar(x + width/2, data['weights of significant forward model - replicated dataset'], width, label='weights of significant forward model - replicated dataset')

# Set the y axis to log scale
ax.set_yscale('symlog')

# Add some text for labels, title, and custom x-axis tick labels, etc.
#ax.set_xlabel('Description', fontsize=14)
ax.set_ylabel('Weights', fontsize=14)
ax.set_title('Comparison of weights of significant models on replicated and test dataset', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(data['Description'], rotation=45, ha='right', fontsize=12)
ax.legend()

# Function to add labels on the bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

#add_labels(bars1)
#add_labels(bars2)

fig.tight_layout()
print('\n')
plt.show()
```
